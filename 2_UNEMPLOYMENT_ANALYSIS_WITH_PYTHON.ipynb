```python
plt.figure(figsize=(10, 10))
sns.histplot(data[' Estimated Unemployment Rate (%)'], kde=True, bins=20, color="skyblue")
plt.title('Distribution of Unemployment Rate(%)')
plt.xlabel('Estinated Unemployment Rate (%)')
plt.ylabel('Frequency')
plt.show()
```

```python
unemployment=data.groupby('Region')[' Estimated Unemployment Rate (%)'].mean()
plt.figure(figsize=(10, 8))
unemployment.plot(kind='barh', color='cyan')
plt.title('Average Unemployment Rate')
plt.xlabel('Average Estimated Unemployment Rate (%)')
plt.ylabel('Region')
plt.show()
```

```python
data = pd.DataFrame(data)
y=data[' Estimated Employed']
x=data['Region']
pr= plt.figure(figsize=(40, 15))
plt.xlabel("States",fontsize = 40)
plt.ylabel("Estimated employment rate",fontsize = 40)
plt.xticks(fontweight='bold',rotation='vertical',fontsize=20)
plt.bar(x,y, color='g',align='center')
plt.show()
```

```python
plt.figure(figsize=(15, 10))
for i,z in enumerate([' Estimated Unemployment Rate (%)',' Estimated Employed', ' Estimated Labour Participation Rate (%)']):
    plt.subplot(2,2,i+1)
    sns.histplot(x=z, hue="Area", data=data, kde=True, palette="hls")
    plt.title("Estimated Employment Rate by Area")
    plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize = (15,6))
sns.boxplot(x='Region', y=' Estimated Unemployment Rate (%)', data=data,)
plt.title('Estimated Unemployment Rate')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize = (15,6))
sns.boxplot(x='Region', y=' Estimated Labour Participation Rate (%)', data=data,)
plt.title('Estimated Labour Participation Rate')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize =(15,8))
data_grouped = data.groupby('Region')[' Estimated Unemployment Rate (%)'].sum()
data_grouped.plot(kind='pie', autopct='%1.1f%%')
plt.title('Estimated Unemployment Rate')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize =  (15,8))
data_grouped = data.groupby('Region')[' Estimated Labour Participation Rate (%)'].sum()
data_grouped.plot(kind='pie', autopct='%1.1f%%')
plt.title('Estimated Labour Participation Rate')
plt.axis('equal')
plt.show()
```
