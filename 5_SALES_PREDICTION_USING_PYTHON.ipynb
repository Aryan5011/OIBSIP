```python
plt.figure(figsize=(4,4))
sns.scatterplot(data=data,x='TV',y='Sales')
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.scatterplot(data=data,x='Radio',y='Sales')
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.scatterplot(data=data,x='Newspaper',y='Sales')
plt.show()
```

```python
plt.figure(figsize = (15,8))
for i,z in enumerate(['TV', 'Radio', 'Newspaper']):
    plt.subplot(2,2,i+1)
    sns.histplot(data = data, x = z,kde = True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=data[['TV', 'Radio', 'Newspaper']])
plt.title('Advertising Costs')
```
